```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset nakshatra.csv')
df = df.drop('source_id',1)
df.head()
```

```python
plt.figure(figsize=(16, 16))
sns.heatmap(df.corr(), cmap="YlGnBu", annot=True, square = True)
plt.title("How are the features Correlated?")
plt.savefig('tempFile.png', dpi=62.5, pad_inches=0)
```
